```python
import pandas as pd
#чтение данных
train = pd.read_csv("train.csv")
members = pd.read_csv("members.csv")
songs = pd.read_csv("songs.csv")
test = pd.read_csv("test.csv")
song_info = pd.read_csv("song_extra_info.csv")
train
```

```python
#далее я объединяю все свои данные в один df
train_data = pd.merge(train, songs, on='song_id', how='left')
test_data = pd.merge(test, songs, on='song_id', how='left')
del songs

train_data.info()
```

```python

train_data = pd.merge(train_data, members, on='msno', how='left')
test_data = pd.merge(test_data, members, on='msno', how='left')
del members

train_data.info()
```

```python

train_data = pd.merge(train_data, song_info, on='song_id', how='left')
test_data = pd.merge(test_data, song_info, on='song_id', how='left')
del song_info

train_data.info()
```

```python
#хочу избавиться от nullов и оценить, сколько различных значений принимает каждое поле
def check_values(data):
    res = pd.DataFrame()
    res["Null values"]=data.isnull().sum()
    res["Unique values"]=data.nunique().sort_values(ascending=True)
    return res

table_value = check_values(train_data)
table_value
```

```python
import numpy as np
#избавляюсь от пустых значений.
def delete_nans(data):
    categories = data.select_dtypes(include='O').columns
    numbers = data.select_dtypes(include=(np.number)).columns
    for col in categories:
        data[col] = data[col].fillna('Unknown')
    for col in numbers:
        data[col] = data[col].interpolate(method='linear')
    return data
```

```python
train_data = delete_nans(train_data)
table_value = check_values(train_data)
table_value
```

```python
#представлю даты как набор числовых полей для повышения колличества признаков
def fix_time(data):
    dates = ['registration_init_time','expiration_date']
    for date in dates:
        data[date] = pd.to_datetime(data[date], format='%Y%m%d', errors='ignore')
        data[date[:3] + '_year'] = data[date].dt.year
        data[date[:3] + '_month'] = data[date].dt.month
        data[date[:3] + '_day'] = data[date].dt.day
    data.drop(dates, axis = 1 , inplace= True)
    return data

train_data = fix_time(train_data)
train_data
```

```python
from sklearn.preprocessing import LabelEncoder
#эндокдинг, приведение категориальных признаков к числовым
categories = train_data.select_dtypes(include="O").columns
encoder = LabelEncoder()
for category in categories:
    train_data[category] = encoder.fit_transform(train_data[category])
```

```python
train_data.drop(['msno', 'song_id'], axis = 1 , inplace= True)
train_data
```
